```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches

%matplotlib inline

plt.rcParams['figure.dpi']= 150
#plt.rcParams['figure.figsize'] = [5,5]
```

```python
# construct the lines that make up the feasible region
x1 = np.linspace(0,20,2000)
x21 = (1.8-0.2*x1)/0.3
x22 = (1.2-0.2*x1)/0.1
x23 = (2.4-0.3*x1)/0.3
```

```python
# point x1=8, x2=8 is not feasible because it does not satisfy the constraints
plt.plot(8,8,'ro')
plt.xlim((-.1,10))
plt.ylim((-.1,12))
plt.gca().set_aspect('equal', adjustable='box')
plt.grid()
plt.xlabel(r'$x_1$ [t]')
plt.ylabel(r'$x_2$ [t]')
plt.savefig('fig1.png', bbox_inches='tight')
```

```python
# point (2,2) is not optimal because we are left with unsatisfied demand
plt.plot(2,2,'ro')
plt.xlim((-.1,10))
plt.ylim((-.1,12))
plt.grid()
plt.gca().set_aspect('equal', adjustable='box')

plt.xlabel(r'$x_1$ [t]')
plt.ylabel(r'$x_2$ [t]')
plt.savefig('fig2.png', bbox_inches='tight')
```

<h2>How to draw the feasible region of an LP problem</h2>

```python
# we display the feasible region with respect to specific constraints as a grey area in the graph
# here the all the points in the graph where x1 >= 0 and x2 >= 0 are feasible, if we do not take into account the other constraints
plt.xlim((-.1,10))
plt.ylim((-.1,12))
plt.gca().set_aspect('equal', adjustable='box')
plt.grid()
plt.xlabel(r'$x_1$ [t]')
plt.ylabel(r'$x_2$ [t]')

plt.annotate('', xy=(1,1), xytext=(1,0), arrowprops=dict(facecolor='black', shrink=0.05))
plt.annotate('', xy=(3,1), xytext=(3,0), arrowprops=dict(facecolor='black', shrink=0.05))
plt.annotate('', xy=(5,1), xytext=(5,0), arrowprops=dict(facecolor='black', shrink=0.05))
plt.annotate('', xy=(7,1), xytext=(7,0), arrowprops=dict(facecolor='black', shrink=0.05))
plt.annotate('', xy=(9,1), xytext=(9,0), arrowprops=dict(facecolor='black', shrink=0.05))

plt.annotate('', xy=(.6,1), xytext=(0,1), arrowprops=dict(facecolor='black', shrink=0.05))
plt.annotate('', xy=(.6,3), xytext=(0,3), arrowprops=dict(facecolor='black', shrink=0.05))
plt.annotate('', xy=(.6,5), xytext=(0,5), arrowprops=dict(facecolor='black', shrink=0.05))
plt.annotate('', xy=(.6,7), xytext=(0,7), arrowprops=dict(facecolor='black', shrink=0.05))
plt.annotate('', xy=(.6,9), xytext=(0,9), arrowprops=dict(facecolor='black', shrink=0.05))
plt.annotate('', xy=(.6,11), xytext=(0,11), arrowprops=dict(facecolor='black', shrink=0.05))
plt.fill_between(x1, 12, color='grey', alpha=0.5)

plt.savefig('fig3.png', bbox_inches='tight')
```

```python
# here the feasible region is constrained by the nonnegativity constraints and the French fries maximum demand
plt.xlim((-.1,10))
plt.ylim((-.1,12))
plt.plot(x1,x21,label=r'$0.2x_1 + 0.3x_2 \leq 1.8$')
plt.annotate('French fries', xy=(3.5,3.5), xytext=(4,5), arrowprops=dict(facecolor='black', shrink=0.05))
plt.gca().set_aspect('equal', adjustable='box')

plt.grid()
plt.xlabel(r'$x_1$ [t]')
plt.ylabel(r'$x_2$ [t]')
plt.fill_between(x1, x21, color='grey', alpha=0.5)

plt.savefig('fig4.png', bbox_inches='tight')
```

```python
# here we add the hash brown max demand into a constraint
plt.xlim((-.1,10))
plt.ylim((-.1,12))
plt.plot(x1,x21,label=r'$0.2x_1 + 0.3x_2 \leq 1.8$')
plt.plot(x1,x22,label=r'$0.2x_1 + 0.1x_2 \leq 1.2$')
plt.annotate('Hash browns', xy=(3,6), xytext=(4,7), arrowprops=dict(facecolor='black', shrink=0.05))
plt.gca().set_aspect('equal', adjustable='box')

plt.grid()
plt.xlabel(r'$x_1$ [t]')
plt.ylabel(r'$x_2$ [t]')
plt.fill_between(x1, np.minimum(x21,x22), color='grey', alpha=0.5)

plt.savefig('fig5.png', bbox_inches='tight')
```

```python
# and here we add flakes max demand as a constraint but notice that it is outside of the feasible region determined by the other two constraints

plt.plot(x1,x21,label=r'$0.2x_1 + 0.3x_2 \leq 1.8$')
plt.plot(x1,x22,label=r'$0.2x_1 + 0.1x_2 \leq 1.2$')
plt.plot(x1,x23,label=r'$0.3x_1 + 0.3x_2 \leq 2.4$')

plt.annotate('Flakes', xy=(4.5,3.5), xytext=(5,5), arrowprops=dict(facecolor='black', shrink=0.05))
plt.gca().set_aspect('equal', adjustable='box')

plt.xlim((-.1,10))
plt.ylim((-.1,12))
plt.grid()
plt.xlabel(r'$x_1$ [t]')
plt.ylabel(r'$x_2$ [t]')
#plt.legend(bbox_to_anchor=(1.05,1), loc=2, borderaxespad=0)
plt.fill_between(x1, np.minimum(x21,x22), color='grey', alpha=0.5)

plt.savefig('fig6.png', bbox_inches='tight')
```

```python
# now we can see why (8,8) is not feasible and why (2,2) is not optimal
plt.plot(x1,x21,label=r'$0.2x_1 + 0.3x_2 \leq 1.8$ (French fries)')
plt.plot(x1,x22,label=r'$0.2x_1 + 0.1x_2 \leq 1.2$ (Hash browns)')
plt.plot(x1,x23,label=r'$0.3x_1 + 0.3x_2 \leq 2.4$ (Flakes)')
plt.plot(2,2,'ro')
plt.plot(8,8,'ro')
plt.xlim((-.1,10))
plt.ylim((-.1,12))
plt.gca().set_aspect('equal', adjustable='box')

plt.grid()
plt.xlabel(r'$x_1$ [t]')
plt.ylabel(r'$x_2$ [t]')
#plt.legend(bbox_to_anchor=(1.05,1), loc=2, borderaxespad=0)
plt.fill_between(x1, np.minimum(x21,x22), color='grey', alpha=0.5)

plt.savefig('fig7.png', bbox_inches='tight')
```

<h2>How to solve the LP problem graphically by changing the objective function value</h2>

```python
# let us draw the objective function indifference line into the same graph
# the indifference line means that all points in the line have the same objective function value, in this case £12000
x2z = (12000 - 5000*x1)/6000
plt.plot(x1,x21,label=r'$0.2x_1 + 0.3x_2 \leq 1.8$ (French fries)')
plt.plot(x1,x22,label=r'$0.2x_1 + 0.1x_2 \leq 1.2$ (Hash browns)')
plt.plot(x1,x23,label=r'$0.3x_1 + 0.3x_2 \leq 2.4$ (Flakes)')
plt.plot(x1,x2z,'k--',label=r'$5000x_1+6000x_2=12000$', linewidth=2)
plt.xlim((-.1,6.1))
plt.ylim((-.1,6.1))
plt.gca().set_aspect('equal', adjustable='box')

plt.grid()
plt.xlabel(r'$x_1$ [t]')
plt.ylabel(r'$x_2$ [t]')
#plt.legend(bbox_to_anchor=(1.05,1), loc=2, borderaxespad=0)
plt.fill_between(x1, np.minimum(x21,x22), color='grey', alpha=0.5)

plt.savefig('fig9.png', bbox_inches='tight')
```

```python
# gradient
x2z = (12000 - 5000*x1)/6000
x2grad = 6*x1/5 # line that is of the form x2 = k*x1, where k = delta y / delta x = 6/5
plt.plot(x1,x21,label=r'$0.2x_1 + 0.3x_2 \leq 1.8$ (French fries)')
plt.plot(x1,x22,label=r'$0.2x_1 + 0.1x_2 \leq 1.2$ (Hash browns)')
plt.plot(x1,x23,label=r'$0.3x_1 + 0.3x_2 \leq 2.4$ (Flakes)')
plt.plot(x1,x2z,'k--',label=r'$5000x_1+6000x_2=12000$', linewidth=2)
#plt.plot(x1,x2grad,'r-')
plt.grid()

plt.arrow(0,0,2.5,3, head_width=0.1, head_length=0.2, fc='r', ec='r')
plt.xlim((0,6))
plt.ylim((0,6))
plt.gca().set_aspect('equal', adjustable='box')
plt.xlabel(r'$x_1$ [t]')
plt.ylabel(r'$x_2$ [t]')
#plt.legend(bbox_to_anchor=(1.05,1), loc=2, borderaxespad=0)
plt.fill_between(x1, np.minimum(x21,x22), color='grey', alpha=0.5)

plt.savefig('fig14.png', bbox_inches='tight')
```

```python
# by increasing the objective function value we see that the indifference line moves away from the origin but parallel with the previous line
x2z = (24000 - 5000*x1)/6000
plt.plot(x1,x21,label=r'$0.2x_1 + 0.3x_2 \leq 1.8$ (French fries)')
plt.plot(x1,x22,label=r'$0.2x_1 + 0.1x_2 \leq 1.2$ (Hash browns)')
plt.plot(x1,x23,label=r'$0.3x_1 + 0.3x_2 \leq 2.4$ (Flakes)')
plt.plot(x1,x2z,'k--',label=r'$5000x_1+6000x_2=24000$', linewidth=2)
plt.xlim((-.1,6.1))
plt.ylim((-.1,6.1))
plt.gca().set_aspect('equal', adjustable='box')

plt.grid()
plt.xlabel(r'$x_1$ [t]')
plt.ylabel(r'$x_2$ [t]')
#plt.legend(bbox_to_anchor=(1.05,1), loc=2, borderaxespad=0)
plt.fill_between(x1, np.minimum(x21,x22), color='grey', alpha=0.5)

plt.savefig('fig10.png', bbox_inches='tight')
```

```python
# by choosing the objective function value such that the indifference line touches an outer point of the feasible region we see that the constraint of hash browns is now fully satisfied 
x2z = (30000 - 5000*x1)/6000
plt.plot(x1,x21,label=r'$0.2x_1 + 0.3x_2 \leq 1.8$ (French fries)')
plt.plot(x1,x22,label=r'$0.2x_1 + 0.1x_2 \leq 1.2$ (Hash browns)')
plt.plot(x1,x23,label=r'$0.3x_1 + 0.3x_2 \leq 2.4$ (Flakes)')
plt.plot(x1,x2z,'k--',label=r'$5000x_1+6000x_2=30000$', linewidth=2)
plt.plot(6,0,'ro')

plt.xlim((-.1,6.1))
plt.ylim((-.1,6.1))
plt.gca().set_aspect('equal', adjustable='box')

plt.grid()
plt.xlabel(r'$x_1$ [t]')
plt.ylabel(r'$x_2$ [t]')
#plt.legend(bbox_to_anchor=(1.05,1), loc=2, borderaxespad=0)
plt.fill_between(x1, np.minimum(x21,x22), color='grey', alpha=0.5)

plt.savefig('fig11.png', bbox_inches='tight')
```

```python
# same as previous but for french fries
x2z = (36000 - 5000*x1)/6000
plt.plot(x1,x21,label=r'$0.2x_1 + 0.3x_2 \leq 1.8$ (French fries)')
plt.plot(x1,x22,label=r'$0.2x_1 + 0.1x_2 \leq 1.2$ (Hash browns)')
plt.plot(x1,x23,label=r'$0.3x_1 + 0.3x_2 \leq 2.4$ (Flakes)')
plt.plot(x1,x2z,'k--',label=r'$5000x_1+6000x_2=36000$', linewidth=2)
plt.plot(0,6,'ro')

plt.xlim((-.1,6.1))
plt.ylim((-.1,6.1))
plt.gca().set_aspect('equal', adjustable='box')

plt.grid()
plt.xlabel(r'$x_1$ [t]')
plt.ylabel(r'$x_2$ [t]')
#plt.legend(bbox_to_anchor=(1.05,1), loc=2, borderaxespad=0)
plt.fill_between(x1, np.minimum(x21,x22), color='grey', alpha=0.5)

plt.savefig('fig12.png', bbox_inches='tight')
```

```python
# now both constraints of french fries and hash browns are satisfied and we find that we cannot increase objective function value anymore without being driven out from the feasible region
# this is the optimal solution and it completes the graphical analysis of an LP problem
x2z = (40500 - 5000*x1)/6000
plt.plot(x1,x21,label=r'$0.2x_1 + 0.3x_2 \leq 1.8$ (French fries)')
plt.plot(x1,x22,label=r'$0.2x_1 + 0.1x_2 \leq 1.2$ (Hash browns)')
plt.plot(x1,x23,label=r'$0.3x_1 + 0.3x_2 \leq 2.4$ (Flakes)')
plt.plot(x1,x2z,'k--',label=r'$5000x_1+6000x_2=40500$', linewidth=2)
plt.plot(4.5,3,'ro')
plt.xlim((-.1,6.1))
plt.ylim((-.1,6.1))
plt.gca().set_aspect('equal', adjustable='box')

plt.grid()
plt.xlabel(r'$x_1$ [t]')
plt.ylabel(r'$x_2$ [t]')
#plt.legend(bbox_to_anchor=(1.05,1), loc=2, borderaxespad=0)
plt.fill_between(x1, np.minimum(x21,x22), color='grey', alpha=0.5)

plt.savefig('fig13.png', bbox_inches='tight')
```

<h2>Illustration of all the feasible points</h2>

```python
# in a graphical solution procedure, it is important to know that the optimal solution will be found on the set of feasible solutions
# the set of feasible solutions are the points on the boundary of the feasible region
# in other words, the red points and all points on the constraint lines between them
x2z = (40500 - 5000*x1)/6000
plt.plot(x1,x21,label=r'$0.2x_1 + 0.3x_2 \leq 1.8$ (French fries)')
plt.plot(x1,x22,label=r'$0.2x_1 + 0.1x_2 \leq 1.2$ (Hash browns)')
plt.plot(x1,x23,label=r'$0.3x_1 + 0.3x_2 \leq 2.4$ (Flakes)')
plt.plot(x1,x2z,'k--',label=r'$5000x_1+6000x_2=40500$', linewidth=2)
plt.plot(4.5,3,'ro')
plt.plot(6,0,'ro')
plt.plot(0,0,'ro')
plt.plot(0,6,'ro')
plt.xlim((-.1,6.1))
plt.ylim((-.1,6.1))
plt.gca().set_aspect('equal', adjustable='box')

plt.grid()
plt.xlabel(r'$x_1$ [t]')
plt.ylabel(r'$x_2$ [t]')
plt.legend(bbox_to_anchor=(1.05,1), loc=2, borderaxespad=0)
plt.fill_between(x1, np.minimum(x21,x22), color='grey', alpha=0.5)
```
