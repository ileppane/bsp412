<p>We first import the necessary components to draw our plot.</p>
<p>The last line ensures that the plot is displayed properly on Jupyter Notebook.</p>

```python
# https://matplotlib.org/users/pyplot_tutorial.html
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

<p>We then construct the arrays that represent the components of the plot that we are drawing. These include the constraint lines and the objective function line.</p>
<p>This step is analogous to an Excel operation where we create numbers in a column in one worksheet. Unlike in Excel, in a programming environment the values of the variables are not 'shown' to the user at all times.</p>
<p>Note that before we can assign values for the y's, we need to create values for the x's. This is done by assigning a linear space (=vector of numbers) value for x.</p>

```python
# Construct lines
# x > 0
x = np.linspace(0, 20, 2000)
# y >= 2
y1 = (x*0) + 2
# 2y <= 25 - x
y2 = (25-x)/2.0
# 4y >= 2x - 8 
y3 = (2*x-8)/4.0
# y <= 2x - 5 
y4 = 2 * x -5
```

<p>Let us see what the plot components look like. The x is a vector just as we wanted, and the y's are vectors based on the values of x just like we wanted.</p>

```python
x
```

<p>We then need the objective function line as well. In this case the line is not uniquely determined as the idea is to vary x and y to obtain different values for the objective function. We thus need to pick some value for the objective function and then see whether it can be displayed at the plot at the area that the contraint lines are populating.</p>

```python
# Objective function line z=4x+3y at z = 50
yz = (50-4*x)/3
```

<p>Now we have all the plot components that we need for drawing.</p>
<p>Plotting can be done using the plot function of Matplotlib.pyplot.</p>

```python
# Make plot
plt.plot(x, y1, label=r'$y\geq2$')
plt.plot(x, y2, label=r'$2y\leq25-x$')
plt.plot(x, y3, label=r'$4y\geq 2x - 8$')
plt.plot(x, y4, label=r'$y\leq 2x-5$')
plt.plot(x, yz, label=r'$z$', linewidth=3, color='k')
```

<p>This looks fairly good as we can see all the lines that we are drawing. However, we need to zoom and label the plot.</p>
<p>Labelling a plot is of utmost importance when presenting results to other people. Please make sure you always label your plots!</p>

```python
# Make plot
plt.plot(x, y1, label=r'$y\geq2$')
plt.plot(x, y2, label=r'$2y\leq25-x$')
plt.plot(x, y3, label=r'$4y\geq 2x - 8$')
plt.plot(x, y4, label=r'$y\leq 2x-5$')
plt.plot(x, yz, label=r'$z$', linewidth=3, color='k')
plt.xlim((0, 16))
plt.ylim((0, 11))
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

<p>Now we know that the feasible region is the area that is circled by the coloured constraint lines. It would be desirable to have it filled with a different colour than white.</p>

```python
# Fill feasible region
plt.plot(x, y1, label=r'$y\geq2$')
plt.plot(x, y2, label=r'$2y\leq25-x$')
plt.plot(x, y3, label=r'$4y\geq 2x - 8$')
plt.plot(x, y4, label=r'$y\leq 2x-5$')
plt.plot(x, yz, label=r'$z$', linewidth=3, color='k')
plt.xlim((0, 16))
plt.ylim((0, 11))
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
y5 = np.minimum(y2, y4)
y6 = np.maximum(y1, y3)
plt.fill_between(x, y5, y6, where=y5>y6, color='grey', alpha=0.5)
```

<p>What we can do next is make use of modularity and gather the plotting into a neat function. We can leave the objective function value z=50 undetermined, so that when calling the function we assign its value.</p>
