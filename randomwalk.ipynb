<h1>One random walk</h1>

```python
# CREATE ONE RANDOM WALK OVER 365 DAYS
import random
walk = [100]
for t in range(1,365):
    pchange = random.choice([-1,1])
    nextprice = walk[t-1]+pchange
    walk.append(nextprice)
    
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(walk)
```

<h1>500 random walks</h1>

```python
# CREATE 500 RANDOM WALKS OVER 365 DAYS
# SAVE THE ENDPOINT OF EACH WALK IN A LIST
endp = []
for mc in range(1,500):
    walk = [100]
    for t in range(1,365):
        pchange = random.choice([-1,1])
        nextprice = walk[t-1]+pchange
        walk.append(nextprice)
    plt.plot(walk)
    endp.append(walk[364])
```

<h1>Distribution of random walk endpoints</h1>

```python
# PLOT THE HISTOGRAM OF THE ENDPOINTS, I.E. THE DISTRIBUTION OF 
# PREDICTED NEW YEAR EVE'S STOCK PRICES
plt.hist(endp)
```
